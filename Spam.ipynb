```python
import pandas as pd
```

```python
df = pd.read_csv('/content/spam ML.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Category'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['Category'] = le.fit_transform(df['Category'])
```

```python
df.drop(columns = ['tr_Cate'],inplace = True)
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
ce = CountVectorizer()
```

```python
data = ce.fit_transform(df['Message']).toarray()
```

```python
feature = ce.get_feature_names_out()
```

```python
len(feature)
```

```python
feature[6000]
```

```python
new_df = pd.DataFrame(data,columns=feature)
```

```python
new_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(new_df,df['Category'])
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
